```python
import pandas as pd

df1 = pd.read_csv("Unemployment in India.csv")
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

print("---- Dataset 1 ----")
print(df1.head(), "\n")

print("---- Dataset 2 ----")
print(df2.head())
```

```python
df1.columns
```

```python
df1.columns = ['Region', 'Date', 'Frequency', 'Estimated_Unemployment_Rate', 'Estimated_Employed', 'Estimated_Labour_Participation_Rate', 'Area']
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1.info()
df1.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

region_unemp = df1.groupby('Region')['Estimated_Unemployment_Rate'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=region_unemp.values, y=region_unemp.index, palette='coolwarm')
plt.title("Average Unemployment Rate by Region (2020)")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Region")
plt.show()
```

```python
# Sort data by date
df1 = df1.sort_values(by='Date')

# Plot unemployment rate trend
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated_Unemployment_Rate', data=df1, color='orange')
plt.title("Unemployment Rate Trend During 2020 (India)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(x='Date', y='Estimated_Unemployment_Rate', hue='Region', data=df1)
plt.title("Region-wise Unemployment Trends (2020)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
